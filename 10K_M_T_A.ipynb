```python
import pandas as pd
```

```python
df = pd.read_csv('10K_M_T_A.csv')
```

```python
df = df.sort_values(['Company', 'Fiscal Year'])
print(df[['Fiscal Year', 'Total Revenue', 'Net Income', 'Company']])
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100
```

```python
print(df[['Fiscal Year', 'Total Revenue', 'Net Income', 'Company', 'Revenue Growth (%)', 'Net Income Growth (%)']])
```

```python
## All the figures are in Millions.
## In this section I have used 'groupby()' and 'pct_change()' functions to find out the revenue growth percentage and net income growth percentage.
```

```python
## Observation:
## For Apple -> we can observe that the fiscal year 2024 had more revenue growth in comparison to 2023, but still had less income growth.
## For Microsoft -> there was a better income growth between 2023 and 2024, in comparison to 2024 and 2025, but overall has been growing over the last three years.
## For Tesla -> it is a loss making company, but performed better in 2025 compared to 2024.
```

```python
df['Net Profit Margin (%)'] = (df['Net Income'] / df['Total Revenue']) * 100
print(df[['Fiscal Year', 'Total Revenue', 'Net Income', 'Company', 'Net Profit Margin (%)']])
```

```python
## Net Profit Margin
## In case of Apple, we can observe that the company is able to have a profit margin which is pretty much steady for the last three years, showing a stagnant but steady business.
## In case of Microsoft, we can see that the company is able to gradually and consistently grow year by year, with the hgihest profits among these three companies.
## In case of Tesla, as we could observe from the above statistics, the company is not able grow their profits, rather its profits have decrease over the years.
```

```python
volatility = df.groupby('Company')['Net Profit Margin (%)'].std()
print(df[['Fiscal Year', 'Total Revenue', 'Net Income', 'Company']] , volatility)
```

```python
## volatility
## Apple: Moderate stability. It has some fluctuations, but it is relatively predictable.
## Microsoft: The most "stable" of the three. This low SD indicates very consistent year-over-year performance.
## Tesla: Extremely high volatility compared to the others. This confirms that Tesla is a high-growth, high-risk company.
```

```python
df['Operating Efficiency (%)'] = (df['Cash Flow'] / df['Total Revenue']) * 100
print(df[['Fiscal Year', 'Total Revenue', 'Net Income', 'Company']])
print()
avg_oe = df.groupby('Company')['Operating Efficiency (%)'].mean()
print("Average Operating Efficiency per Company:")
print(avg_oe)
```

```python
## Average Operating Efficiency
## Apple is highly efficient for a company that sells physical goods, but it faces "hardware gravity."
## Conclusion: Microsoft is the most efficient at generating cash.
## Tesla: A lower number here isn't necessarily "bad"—it reflects an aggressive growth stage where cash is being poured back into the physical business.
```

```python
summary = df.groupby('Company').agg({
    'Revenue Growth (%)': 'mean',
    'Net Income Growth (%)': 'mean'
}).reset_index()

print("\nYear-over-Year Average Growth Rates (%):")
print(summary)
```
